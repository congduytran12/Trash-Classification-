```python
import matplotlib.pyplot as plt
```

```python
ce_train_acc = []
hin_train_acc = []
hin_train_loss = []
ce_train_loss = []
f = open("train_acc_0.01.txt","r")
for i in range(40):
    ce_train_acc.append(float(f.readline()))
f.close()
f = open("hin_train_acc_0.01.txt","r")
for i in range(40):
    hin_train_acc.append(float(f.readline()))
f.close()
f = open("train_loss_0.01.txt","r")
for i in range(40):
    ce_train_loss.append(float(f.readline()))
f.close()
f = open("hin_train_loss_0.01.txt","r")
for i in range(40):
    hin_train_loss.append(float(f.readline()))
f.close()

fig, ax1 = plt.subplots()
ax1.set_xlabel('epochs')
ax1.set_ylabel('accuracy')
ax1.plot(range(1,41),ce_train_acc,color="blue")
ax1.plot(range(1,41),hin_train_acc,color="orange")
ax1.legend(["Model 1 acc.","Model 2 acc."],loc="upper right",bbox_to_anchor=(1, 0.9))
ax2 = ax1.twinx()
ax2.set_ylabel('loss')
ax2.plot(range(1,41),ce_train_loss,color="green")
ax2.plot(range(1,41),hin_train_loss,color="red")
ax2.legend(["Model 1 loss","Model 2 loss"],loc="lower right",bbox_to_anchor=(1, 0.15))

plt.title("Training Loss: Plain Implementation")
plt.savefig("training1.png")
```
